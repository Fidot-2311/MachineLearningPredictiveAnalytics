```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Assuming you have already selected the features and standardized the data

# Select only the selected features for both training and test sets
X_train_selected = X_train[top_n_features]
X_test_selected = X_test[top_n_features]

# Standardize the selected features
scaler = StandardScaler()
X_train_selected_scaled = scaler.fit_transform(X_train_selected)
X_test_selected_scaled = scaler.transform(X_test_selected)
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression

# Set random seed for reproducibility
np.random.seed(42)

# Train the logistic regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_train_selected_scaled, y_train)

# Predictions on the training set
logistic_train_pred = logistic_model.predict(X_train_selected_scaled)

# Predictions on the testing set
logistic_test_pred = logistic_model.predict(X_test_selected_scaled)

# Calculate ROC-AUC score for training set
logistic_train_roc_auc = roc_auc_score(y_train, logistic_train_pred)

# Calculate ROC-AUC score for testing set
logistic_test_roc_auc = roc_auc_score(y_test, logistic_test_pred)

# Print training set metrics
print("Logistic Training Set Metrics:")
print(classification_report(y_train, logistic_train_pred, digits=4))
print("Confusion Matrix (Training Set):")
print(confusion_matrix(y_train, logistic_train_pred))
print("Training Accuracy:", accuracy_score(y_train, logistic_train_pred))
print("ROC-AUC Score (Training Set):", logistic_train_roc_auc)

# Print testing set metrics
print("\nLogistic Testing Set Metrics:")
print(classification_report(y_test, logistic_test_pred, digits=4))
print("Confusion Matrix (Testing Set):")
print(confusion_matrix(y_test, logistic_test_pred))
print("Testing Accuracy:", accuracy_score(y_test, logistic_test_pred))
print("ROC-AUC Score (Testing Set):", logistic_test_roc_auc)
```

```python
# Compute confusion matrix for training set
logistic_train_conf_matrix = confusion_matrix(y_train, logistic_train_pred)

# Compute confusion matrix for testing set
logistic_test_conf_matrix = confusion_matrix(y_test, logistic_test_pred)

# Plot confusion matrix for testing set
plt.figure(figsize=(4, 3))
sns.heatmap(logistic_test_conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Testing Set)')
plt.show()
```
